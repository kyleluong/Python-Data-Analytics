```python
# Why we're utilizing the mean values:
# R-Squared's weakness is that more amounts of predictor variables will lower the accuracy so we condensed the years
# into a mean value also for simplification of the calculation.

states_df['CoalCMean'] = (states_df['CoalC2010'] + states_df['CoalC2011']
                          + states_df['CoalC2012'] + states_df['CoalC2013'] + states_df['CoalC2014'])/5 

states_df['ElecCMean'] = (states_df['ElecC2010'] + states_df['ElecC2011']
                            + states_df['ElecC2012'] + states_df['ElecC2013'] + states_df['ElecC2014'])/5

states_df['FossFuelCMean'] = (states_df['FossFuelC2010'] + states_df['FossFuelC2011'] + 
                              states_df['FossFuelC2012'] + states_df['FossFuelC2013'] + states_df['FossFuelC2014'])/5

states_df['NatGasCMean'] = (states_df['NatGasC2010'] + states_df['NatGasC2011'] + states_df['NatGasC2012'] +
                            states_df['NatGasC2013'] + states_df['NatGasC2014'])/5

states_df['BiomassCMean'] = (states_df['BiomassC2010'] + states_df['BiomassC2011'] + states_df['BiomassC2012'] + 
                            states_df['BiomassC2013'] + states_df['BiomassC2014'])/5

states_df['GeoCMean'] = (states_df['GeoC2010'] + states_df['GeoC2011'] + states_df['GeoC2012'] + states_df['GeoC2013'] + 
                        states_df['GeoC2014'])/5
                    
    
states_df['HydroCMean'] = (states_df['HydroC2010'] + states_df['HydroC2011'] + states_df['HydroC2012']
                           + states_df['HydroC2013'] + states_df['HydroC2014'])/5

states_df['LPGCMean'] = (states_df['LPGC2010'] + states_df['LPGC2011'] + states_df['LPGC2012'] + 
                        states_df['LPGC2013'] + states_df['LPGC2014'])/5
```

```python
# We are utilizing the median values because it's rational and logically sound as opposed to our initial thought of
# using the mean. There is a trend of the coal consumption mean in the USA decreasing throughout the years and it would
# most definitely skew the data if we used mean values in the linear regression model.


# Could not figure out how to take ROW-WISE Median for Coal consumption in states_df so I copied the columns into a new df
# and then calculated the median from there. Then, I added the median back into states_df
CoalC_df = states_df[['CoalC2010', 'CoalC2011', 'CoalC2012', 'CoalC2013', 'CoalC2014']].copy()
CoalC_df['CoalCMedian'] = CoalC_df.median(axis=1)

ElecC_df = states_df[['ElecC2010', 'ElecC2011', 'ElecC2012', 'ElecC2013', 'ElecC2014']].copy()
ElecC_df['ElecCMedian'] = ElecC_df.median(axis=1)

FossFuelC_df = states_df[['FossFuelC2010', 'FossFuelC2011', 'FossFuelC2012', 'FossFuelC2013', 'FossFuelC2014']].copy()
FossFuelC_df['FossFuelCMedian'] = FossFuelC_df.median(axis=1)

NatGasC_df = states_df[['NatGasC2010', 'NatGasC2011', 'NatGasC2012', 'NatGasC2013', 'NatGasC2014']].copy()
NatGasC_df['NatGasCMedian'] = NatGasC_df.median(axis=1)

BiomassC_df = states_df[['BiomassC2010', 'BiomassC2011', 'BiomassC2012', 'BiomassC2013', 'BiomassC2014']].copy()
BiomassC_df['BiomassCMedian'] = BiomassC_df.median(axis=1)

HydroC_df = states_df[['HydroC2010', 'HydroC2011', 'HydroC2012', 'HydroC2013', 'HydroC2014']].copy()
HydroC_df['HydroCMedian'] = HydroC_df.median(axis=1)

LPGC_df = states_df[['LPGC2010', 'LPGC2011', 'LPGC2012', 'LPGC2013', 'LPGC2014']].copy()
LPGC_df['LPGCMedian'] = LPGC_df.median(axis=1)

GeoC_df = states_df[['GeoC2010', 'GeoC2011', 'GeoC2012', 'GeoC2013', 'GeoC2014']].copy()
GeoC_df['GeoCMedian'] = GeoC_df.median(axis=1)

states_df['CoalCMedian'] = CoalC_df['CoalCMedian']
states_df['ElecCMedian'] = ElecC_df['ElecCMedian']
states_df['FossFuelCMedian'] = FossFuelC_df['FossFuelCMedian']
states_df['NatGasCMedian'] = NatGasC_df['NatGasCMedian']
states_df['BiomassCMedian'] = BiomassC_df['BiomassCMedian']
states_df['HydroCMedian'] = HydroC_df['HydroCMedian']
states_df['LPGCMedian'] = LPGC_df['LPGCMedian']
states_df['GeoCMedian'] = GeoC_df['GeoCMedian']

states_df.head()
```

```python
# The variables that seem to create the most impact on a beneficial model would be the coast, great lakes, region,
# and energy consumption values. However, not all energy resources have collinearity. In this case, Fossil Fuel and 
# Natural Gas are redundant in the predictor variables.

X = states_df[['Coast', 'Great_Lakes', 'Region_1', 'Region_2', 'Region_3',
               'Region_4', 'CoalCMedian', 'BiomassCMedian', 'GeoCMedian', 'HydroCMedian', 'LPGCMedian']]

y = states_df[['ElecCMedian']]

from sklearn.model_selection import train_test_split

# Split data into 80% training, 20% testing.
# Split X and y where X has our predictors and y has our response variable
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)

# Train the model using the training data.
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
lm.fit(X_train, y_train)


# Use .coef method to get the coefficients of each predictor variable. 
#Coefficients tell us the strength of the relationship between the predictor and response (also controls for other predictors)
for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, lm.coef_[0][idx]))
    
    
#.intercept_ method returns an array of intercepts
intercept = lm.intercept_[0]
print("\nThe intercept for our model is {}".format(intercept))    
```

```python
#look at R-squared value for the model
lm.score(X_test, y_test)
```

```python
#next we looked at Mean Squared Error (MSE)
#the models's .predict method predicts responses based on X_test data passed in.  
from sklearn.metrics import mean_squared_error

y_predict = lm.predict(X_test)

#mean_squared_error will look at error by comparing our predicted values to the actual outcomes (y_test)
regression_model_mse = mean_squared_error(y_predict, y_test)

regression_model_mse
```

    <h3>Data Visualization:</h3>
    <p>How do we visualize a multivariable regression model when there's multiple predictor variables? There are a couple of solutions.. We could show a simple linear regression model for each predictor variable vs. the response variable like the one shown below to see how each predictor variable affects the response variable but the problem is that there are too many predictor variables and too many models in which we are trying to predict electricity, coal, fossil fuel, and etc. There would be absolutely way too many graphs and it would take some time required for computation so our solution is to condense each model's visualization into one graph utilizing a dimension reduction technique called Principal Components Analysis (PCA).</p>

```python
# There's so many other predictor variables, so we're not going to graph like this, we'll use PCA instead!
sns.lmplot(x="BiomassCMedian", y="ElecCMedian", data=states_df, size=2.5);
sns.lmplot(x="HydroCMedian", y="ElecCMedian", data=states_df, size=2.5);
sns.lmplot(x="CoalCMedian", y="ElecCMedian", data=states_df, size=2.5);
```

Regions 1 through 3's slopes are relatively near identical whereas region 4's slope is very steep in comparison! This is definitely because of the outliers which happen to be in Region 4. Therefore, a hypothesis could be made that Region 4 consumes the most electricity.<p>

```python
features = ['Coast', 'Great_Lakes', 'Region_1', 'Region_2', 'Region_3','Region_4', 'CoalCMedian', 'BiomassCMedian',
            'GeoCMedian', 'HydroCMedian', 'LPGCMedian']
x = states_df.loc[:, features].values

from sklearn.decomposition import PCA
y = states_df.loc[:,['ElecCMedian']].values
pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents
             , columns = ['PrincipalComponent1', 'PrincipalComponent2'])

principalDf = pd.concat([principalDf, states_df[['Region']]], axis = 1)

sns.lmplot(x="PrincipalComponent1", y="PrincipalComponent2", hue="Region", data=principalDf,
          palette="Set1", size=5);
```
